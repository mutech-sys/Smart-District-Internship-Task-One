```python
raw_products = [
    (" Laptop ", "Electronics", "1200"),
    ("Mouse", "Electronics", "25"),
    ("Keyboard", "Electronics", "45"),
    ("Laptop", "Electronics", "1200"),
    ("Chair", "Furniture", "150"),
    ("Desk", "Furniture", "300"),
    ("Notebook", "Stationery", "5"),
    ("Pen", "Stationery", None),
]
```

```python
catalogue = {}      # dict
seen = set()
```

```python
for name, category, price in raw_products:

    name = name.strip()     # remove all the whitespaces

    if price is None:       # if price is missing skip the data
        continue

    price = float(price)        # convert price to float

    # remove duplicates
    key = (name, category, price)
    if key in seen:
        continue
    seen.add(key)

    if category not in catalogue:       # creating categories for the first time
        catalogue[category] = []

    catalogue[category].append((name, price))       # adding items by categories
```

```python
# print summary
total = 0
for category, products in catalogue.items():

    print(f"\nCategory: {category}")
    print("-" * 30)

    sub_total = 0

    for name, price in products:
        print(f"{name:10}  ${price:.2f}")
        sub_total += price
        total += sub_total

    print(f"Total items: {len(products)}")
    print(f"Total value: ${sub_total:.2f}")

print("-" * 30)
print(f"\nTotal values of all the items: {total:.2f}")
```
