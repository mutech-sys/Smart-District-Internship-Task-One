```python
from dotenv import load_dotenv
import os

load_dotenv()

SOURCE1_KEY = os.getenv("SOURCE1_KEY")
SOURCE2_KEY = os.getenv("SOURCE2_KEY")
SOURCE3_KEY = os.getenv("SOURCE3_KEY")
```

```python
import asyncio
import random

async def fetch_source1():
    await asyncio.sleep(2)
    return {"source": "A", "value": random.randint(10,50)}

async def fetch_source2():
    await asyncio.sleep(3)
    return {"source": "B", "value": random.randint(20,60)}

async def fetch_source3():
    await asyncio.sleep(1)
    return {"source": "C", "value": random.randint(5,40)}
```

```python
async def fetch_all_sources():

    results = await asyncio.gather(
        fetch_source1(),
        fetch_source2(),
        fetch_source3()
    )

    return results
```

```python
from multiprocessing import Pool

def process_record(record):
    value = record["value"]
    return value * 2


def process_data(data):

    with Pool(3) as pool:
        results = pool.map(process_record, data)

    return results
```

```python
import threading
import time

def background_logger():

    for i in range(3):
        print("Logging system status...")
        time.sleep(2)

def start_logger():

    t = threading.Thread(target=background_logger)
    t.start()
    return t
```

```python
def display_summary(raw, processed):

    total_raw = sum(r["value"] for r in raw)
    total_processed = sum(processed)

    print("\n===== DATA SYNC DASHBOARD =====")

    for r in raw:
        print(f"Source {r['source']} -> {r['value']}")

    print("\nMerged Total:", total_raw)
    print("Processed Total:", total_processed)

    print("Records:", len(raw))
```

```python
import asyncio

async def main():

    logger_thread = start_logger()

    print("Fetching data from sources...")

    raw_data = await fetch_all_sources()

    processed = process_data(raw_data)

    display_summary(raw_data, processed)

    logger_thread.join()


await main()
```
