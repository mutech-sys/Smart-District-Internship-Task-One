```python
import os
import csv
```

```python
FOLDER_PATH = "C:\\Users\\muabd\\OneDrive\\Desktop\\SDI\\student_records"     # path where all the student data exists
PASS_MARK = 50      # minimum passing marks
```

```python
clean_results = []      
```

```python
for filename in os.listdir(FOLDER_PATH):
    file_path = os.path.join(FOLDER_PATH, filename)

    # process only CSV files
    if not filename.endswith(".csv"):
        continue

    try:
        with open(file_path, "r", newline="") as f:
            reader = csv.DictReader(f)
            for row in reader:
                try:
                    # get student info
                    name = row.get("Name", "").strip()
                    roll = row.get("Roll", "").strip()
                    marks_str = row.get("Marks", "").strip()

                    # skip if any field is missing
                    if not name or not roll or not marks_str:
                        continue

                    # convert marks to float
                    marks = float(marks_str)

                    # flag failed students
                    status = "Failed" if marks < PASS_MARK else "Passed"

                    # add cleaned result
                    clean_results.append({
                        "Name": name,
                        "Roll": roll,
                        "Marks": marks,
                        "Status": status
                    })

                except ValueError:
                    # handle invalid marks
                    print(f"Invalid marks in file {filename} for student {row.get('Name', 'Unknown')}")
                    continue

    except (FileNotFoundError, PermissionError, csv.Error) as e:
        # handle corrupted or missing file
        print(f"Skipping file {filename} due to error: {e}")
        continue
```

```python
output_file = "clean_results_task_2.csv"
with open(output_file, "w", newline="") as f:
    fieldnames = ["Name", "Roll", "Marks", "Status"]
    writer = csv.DictWriter(f, fieldnames=fieldnames)
    writer.writeheader()
    for result in clean_results:
        writer.writerow(result)

print(f"Processing complete. Clean results saved to {output_file}")
```
