```python
import os
from datetime import datetime, timedelta
```

```python
def scan_folder(folder_path):
    files_info = []     # list of dicts
    now = datetime.now()
    
    for item in os.listdir(folder_path):
        item_path = os.path.join(folder_path, item)
        
        if os.path.isfile(item_path):
            stat_info = os.stat(item_path)
            last_modified = datetime.fromtimestamp(stat_info.st_mtime)
            flagged = (now - last_modified).days > 30
            
            file_info = {
                "name": item,
                "size": stat_info.st_size,
                "type": os.path.splitext(item)[1],
                "last_modified": last_modified,
                "flagged": flagged
            }
            files_info.append(file_info)
    return files_info
```

```python
def generate_audit_report(files_info, report_folder):
    if not os.path.exists(report_folder):
        os.mkdir(report_folder)
        
    timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
    report_path = os.path.join(report_folder, f"audit_report.txt")
    
    with open(report_path, "w") as f:
        f.write("File Audit Report\n")
        f.write(f"Generated: {datetime.now()}\n")
        f.write("="*50 + "\n\n")
        for file in files_info:
            f.write(f"Name: {file['name']}\n")
            f.write(f"Size: {file['size']} bytes\n")
            f.write(f"Type: {file['type']}\n")
            f.write(f"Last Modified: {file['last_modified'].strftime('%Y-%m-%d %H:%M:%S')}\n")
            f.write(f"Flagged (>30 days old): {'YES' if file['flagged'] else 'NO'}\n")
            f.write("-"*40 + "\n")
    return report_path
```

```python
def generate_penalties_report(files_info, report_folder, fine_per_file=50):
  
    if not os.path.exists(report_folder):
        os.mkdir(report_folder)
        
    timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
    report_path = os.path.join(report_folder, f"penalties_report.txt")
    
    with open(report_path, "w") as f:
        f.write("Penalties Report\n")
        f.write(f"Generated: {datetime.now()}\n")
        f.write("="*50 + "\n\n")
        total_fine = 0
        for file in files_info:
            if file['flagged']:
                f.write(f"File: {file['name']} - Fine: ${fine_per_file}\n")
                total_fine += fine_per_file
        f.write("\n" + "="*50 + "\n")
        f.write(f"Total Fine: ${total_fine}\n")
    return report_path
```

```python
project_folder = "C:\\Users\\muabd\\OneDrive\\Desktop\\SDI\\Task 3\\Projects"
report_folder = os.path.join(project_folder, "reports")

files_info = scan_folder(project_folder)

# files_info is a list of dicts

audit_path = generate_audit_report(files_info, report_folder)
print(f"Audit report generated at: {audit_path}")

penalties_path = generate_penalties_report(files_info, report_folder)
print(f"Penalties report generated at: {penalties_path}")

```
