```python
products = [
    {"name": "Laptop", "category": "Electronics", "price": 1200},
    {"name": "Mouse", "category": "Electronics", "price": 25},
    {"name": "Laptop", "category": "Electronics", "price": 1200},   # duplicate
    {"name": "Desk", "category": "Furniture", "price": 300},
    {"name": "Chair", "category": "Furniture", "price": None},      # missing price
    {"name": "Phone", "category": "Electronics", "price": 800},
    {"name": "Sofa", "category": "Furniture", "price": 900},
    {"name": "Table", "category": "Furniture", "price": 400},
    {"name": "Headphones", "category": "Electronics", "price": 150},
    {"name": "Lamp", "category": "Furniture", "price": 120},
]
```

```python
clean_products = [p for p in products if p["price"] is not None]
print(clean_products)
```

```python
unique_products = { (p["name"], p["category"], p["price"]) for p in clean_products}

unique_products = [
    {"name": n, "category": c, "price": pr}
    for (n, c, pr) in unique_products
]

print(unique_products)
```

```python
categories = {p["category"] for p in unique_products}
print(categories)
```

```python
grouped = {
    cat: [p for p in unique_products if p["category"] == cat]
    for cat in categories
}
print(grouped)
```

```python
top5 = {
    cat: sorted(items, key=lambda x: x["price"], reverse=True)[:5]
    for cat, items in grouped.items()
}
print(top5)
```

```python
summary = {
    cat: {
        "total_items": len(items),
        "most_expensive": items[0]["name"] if items else None,
        "average_price": sum(p["price"] for p in items) / len(items)
    }
    for cat, items in top5.items()
}
print(summary)
```
