```python
employees = [
    {"name": "Ali", "score": 90, "experience": 5},
    {"name": "Sara", "score": 95, "experience": 3},
    {"name": "Ahmed", "score": 90, "experience": 7},
    {"name": "Bilal", "score": 95, "experience": 5},
    {"name": "Zara", "score": 90, "experience": 5}
]
```

```python
ranked = sorted(employees, key=lambda emp: (-emp["score"], -emp["experience"], emp["name"])
) 
print(ranked)
```

```python
employees_by_name = sorted(employees, key=lambda emp: emp["name"])
names = [emp["name"] for emp in employees_by_name]
print(names)
```

```python
def binary_search(names, target):

    left = 0
    right = len(names) - 1

    while left <= right:

        mid = (left + right) // 2

        if names[mid] == target:
            return mid

        elif names[mid] < target:
            left = mid + 1

        else:
            right = mid - 1

    return -1
```

```python
target = "Sara"

index = binary_search(names, target)

if index != -1:
    print(employees_by_name[index])
else:
    print("Employee not found")
```
