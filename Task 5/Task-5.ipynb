```python
orders = [
    {"customer": "Ali", "category": "electronics", "price": 1200},
    {"customer": "Sara", "category": "clothing", "price": 200},
    {"customer": "Ahmed", "category": "electronics", "price": 800},
    {"customer": "Fatima", "category": "books", "price": 150},
    {"customer": "Zain", "category": "clothing", "price": 450}
]
```

```python
MIN_ORDER = 300
```

```python
discount_rules = {
    "electronics": 0.10,
    "clothing": 0.20,
    "books": 0.05
}
```

```python
discounted_orders = list(map(
    lambda o: {
        "customer": o["customer"],
        "category": o["category"],
        "original_price": o["price"],
        "final_price": o["price"] * (1 - discount_rules.get(o["category"], 0))}, orders))

for order in discounted_orders:
    print(order)   
```

```python
valid_orders = list(filter(lambda o: o["final_price"] >= MIN_ORDER, discounted_orders))

for order in valid_orders:
    print(order)
```

```python
sorted_orders = sorted(valid_orders, key=lambda o: o["final_price"], reverse=True)

for order in sorted_orders:
    print(order)
```

```python
from functools import reduce

total_revenue = reduce(lambda total, o: total + o["final_price"], sorted_orders, 0)

print(total_revenue)
```
