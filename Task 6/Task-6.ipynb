```python
from itertools import chain
```

```python
# 1. Stream log file line by line
def read_logs(filepath):
    with open(filepath, "r") as f:
        for line in f:
            yield line.strip()
```

```python
# 2. Yield only CRITICAL errors
def critical_errors(lines):
    for line in lines:
        if "CRITICAL" in line:
            yield line
```

```python
# 3. Maintain live running count of error types
def live_counter(errors):
    counts = {}

    for error in errors:
        counts[error] = counts.get(error, 0) + 1
        yield error, counts[error]
```

```python
# 4. Combine multiple log files if needed
logs = chain(
    read_logs("server1.log"),
    read_logs("server2.log")
)
```

```python
# 5. Filter critical errors
critical_stream = critical_errors(logs)
```

```python
error_types = (
    line.split("CRITICAL ")[1]
    for line in critical_stream
)
```

```python
# 7. Show live running count
for error, count in live_counter(error_types):
    print(f"{error} -> {count}")
```
