```python
class BankAccount:
    
    bank_name = "Secure Bank"

    def __init__(self, owner, balance):
        self.owner = owner
        self.__balance = balance   

    @property
    def balance(self):
        return self.__balance

    @balance.setter
    def balance(self, value):
        if value < 0:
            print("Balance cannot be negative.")
        else:
            self.__balance = value

    def deposit(self, amount):
        if self.is_valid_amount(amount):
            self.__balance += amount
            print(f"Deposited {amount}")
        else:
            print("Invalid deposit amount")

    def withdraw(self, amount):
        if not self.is_valid_amount(amount):
            print("Invalid withdrawal amount")
        elif amount > self.__balance:
            print("Insufficient funds")
        else:
            self.__balance -= amount
            print(f"Withdrew {amount}")

    # static method (validation helper)
    @staticmethod
    def is_valid_amount(amount):
        return amount > 0

    # class method
    @classmethod
    def bank_info(cls):
        return f"Welcome to {cls.bank_name}"

    # readable output
    def __str__(self):
        return f"Account Owner: {self.owner}, Balance: {self.__balance}"

    # developer representation
    def __repr__(self):
        return f"BankAccount(owner='{self.owner}', balance={self.__balance})"
```

```python
acc1 = BankAccount("Ali", 1000)

print(BankAccount.bank_info())

acc1.deposit(500)
acc1.withdraw(300)

print(acc1.balance)

print(acc1)
print(repr(acc1))
```
